```python
# 📌 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

from keras.models import Model
from keras.layers import Input, Dense
from keras import regularizers
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')

```

```python
# 📌 2. Load Dataset
df = pd.read_csv('/content/merged_dataset.csv')   # ⬅️ Change path accordingly

print(df.shape)
df.head()
```

```python
# 📌 3. Preprocessing
# Encode categorical columns
categorical_cols = ['protocol_type', 'service', 'flag', 'class']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Separate features and labels
X = df.drop('class', axis=1)
y = df['class']

print(X.shape, y.shape)

# Scale features to [0,1]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 📌 4. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# 📌 5. Build Autoencoder
input_dim = X_train.shape[1]
encoding_dim = 20  # you can tune this

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu", activity_regularizer=regularizers.l1(10e-5))(input_layer)
decoder = Dense(input_dim, activation="sigmoid")(encoder)
autoencoder = Model(inputs=input_layer, outputs=decoder)

autoencoder.compile(optimizer='adam', loss='mse')
```

```python
# 📌 6. Train Autoencoder
history = autoencoder.fit(X_train, X_train,
                          epochs=50,
                          batch_size=128,
                          shuffle=True,
                          validation_data=(X_test, X_test),
                          verbose=1)
```

```python
# 📌 7. Reconstruction Errors
X_pred = autoencoder.predict(X_scaled)
reconstruction_error = np.mean(np.square(X_scaled - X_pred), axis=1)
```

```python
# 📌 8. Find Best Threshold using ROC Curve (Corrected)
from sklearn.metrics import roc_curve, auc

def find_best_threshold(y_true, reconstruction_error):
    # ✅ First, correctly map y_true to binary
    # Assuming 0 = normal, others = attack
    y_binary = np.where(y_true == 0, 0, 1)

    # ✅ Then calculate ROC curve
    fpr, tpr, thresholds = roc_curve(y_binary, reconstruction_error)

    # ✅ Find best threshold
    j_scores = tpr - fpr
    best_idx = j_scores.argmax()
    best_threshold = thresholds[best_idx]
    print(f"Best Threshold: {best_threshold:.6f}")

    # ✅ Plot ROC Curve
    plt.figure(figsize=(8,6))
    plt.plot(fpr, tpr, label=f"AUC = {auc(fpr, tpr):.4f}")
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid()
    plt.show()

    return best_threshold, y_binary  # ⬅️ returning both

# ✅ Usage:
best_threshold, y_binary = find_best_threshold(y, reconstruction_error)
```

```python
# 📌 9. Predict Anomalies
y_pred = (reconstruction_error > best_threshold).astype(int)
```

```python
# 📌 10. Evaluation (Corrected)
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def evaluate_model(y_true, y_pred, model_name="Model"):
    # ✅ Convert y_true to binary labels
    # Assuming 0 = normal, others = anomaly
    y_binary = np.where(y_true == 0, 0, 1)

    print(f"\n🔵 Evaluation Results for {model_name}")
    print(classification_report(y_binary, y_pred, target_names=["Normal", "Anomalous"]))

    cm = confusion_matrix(y_binary, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'{model_name} - Confusion Matrix')
    plt.show()

    acc = np.mean(y_binary == y_pred)
    error_rate = 1 - acc
    print(f"Accuracy: {acc:.4f}")
    print(f"Error Rate: {error_rate:.4f}")

# ✅ Usage:
evaluate_model(y, y_pred, "Autoencoder with Best Threshold")
```
